# Importing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1.Read the dataset to the python environment

```python
df = pd.read_csv("F:\\pythonprogramming\\cars_data.csv")
df
```

# 2.Check for the null values present in the dataset

```python
df.isnull().sum()
```

# 3.Plot a bar graph of male vs female buyers participated in the sales


```python
x=['Male','Female']
#Finding the count of male and female buyers
y=[len(df[df["Buyer Gender"]=='Male']),len(df[df["Buyer Gender"]=='Female'])]
plt.figure(figsize=(5,5))
#ploting the male and female count to the bar chart
plt.bar(x,y)
plt.title("Gender vs Sales",fontsize=16,color='red')
plt.xlabel("Gender",fontsize=14,color='red')
plt.ylabel("Sales",fontsize=14,color='red')
plt.plot
#To display the total male and female buyers count on the top of each bar
for i, j in enumerate(y):
    plt.annotate(str(j),(i,j+1))
    
plt.show()
```

# 4.Find the top 5 cars based on their sales price.

```python
#sorting out car sales price value in ascending order
df_1=df.sort_values(by=['Sale Price'],ascending=False)
df_2=df_1.head(5)
df_2["Make"]
```

# 5. Find the least 5 cars based on their Resell price.


```python
#sorting out car resell price value in descending order
df_3=df.sort_values(by=['Resell Price'],ascending=True)
df_3.head(5)
df_4=df_3.head(5)
df_4["Make"]
```
