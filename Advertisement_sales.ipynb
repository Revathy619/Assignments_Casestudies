# 2.Build a regression model to predict the profit from the given data and comment your findings on it

```python
data.columns
```

```python
#Building a model with profit as Target and five Adds as features
data6= data.drop(['Date'], axis=1)
y= data6['Profit(in Thousands)']
x=data6.drop(['Profit(in Thousands)'],axis=1)
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,random_state=42,test_size=0.25)
from sklearn import linear_model
lr = linear_model.LinearRegression()
model= lr.fit(x_train,y_train)
predictions = model.predict(x_test) 
print(y_test)
print(predictions)
```

```python
#To find the efficiency of our regression model by finding MSE
from sklearn.metrics import mean_squared_error
print("MSE is :",mean_squared_error(y_test,predictions))
```

 MSE value is 8.11273293080854e-25 which is very low, so that our model is perfect

```python
#To find the efficiency of our regression model by finding R^2 Value
from sklearn.metrics import r2_score
r2=r2_score(y_test,predictions)
print("R^2 value is:",r2)
```

Also, R squared value of 1.0 indicates the model is perfect.
